```python
import numpy as np
import pandas as pd
from scipy.signal import blackman, hann, hamming
from numpy.fft import rfft, rfftfreq
import matplotlib.pyplot as plt

```

```python
freq_in_phase = 0.2
freq_out_of_phase = 0.2

t = np.arange(0,2*np.pi/freq_in_phase)
sin_wave_out_of_phase = np.cos(freq_out_of_phase*t) 
sin_wave_in_phase = np.sin(freq_in_phase*t)

accumulated_signal = sin_wave_out_of_phase + sin_wave_in_phase

plt.figure(2)
plt.plot(t, sin_wave_out_of_phase, "r",  label=' out of phase' )
plt.plot(t, sin_wave_in_phase, "g", label = " in phase " )
plt.plot(t, accumulated_signal, "b", label = " in phase " )
plt.show()
#plt.plot(t, sin_wave_in_phase)
```

```python
psd_in = abs(rfft(sin_wave_in_phase))
imag_ampl_in = np.fft.fft(sin_wave_in_phase)
freqs_in = rfftfreq(len(t))
#print(freqs_in)

psd_out = abs(rfft(sin_wave_out_of_phase))
imag_ampl_out = np.fft.fft(sin_wave_out_of_phase)
freqs_out = rfftfreq(len(t))

plt.figure(3)
plt.stem(psd_out[:10], label=' out of phase')
plt.xlabel("First 10 positive frequencies")
plt.ylabel("Spectral density")
plt.legend()

plt.figure(4)
plt.stem(psd_in[0:10], label = " in phase ")
plt.xlabel("First 10 positive frequencies")
plt.ylabel("Spectral density")
plt.legend()
plt.show()
#max_freq_widw = 2.25#freqs_out_widw[psd_out_widw.idxmax()[0]]
#print('max freq', max_freq_widw)


#plt.figure()
#plt.stem(ordered_freq_out[:10], ordered_ampl_out[:10])
#plt.plot(freqs_out, np.absolute(imag_ampl_out))
#plt.plot(freqs_out_widw[:10], psd_out_widw[0][:10])
#plt.legend()
```

```python
N = len(freqs_in)
t_extend = np.arange(0,3*np.pi/freq_in_phase*100)
restored_signal_in = np.zeros(len(t_extend))
restored_signal_out = np.zeros(len(t_extend))

for i in range(N):
    ampli = np.absolute(imag_ampl_out[i]) / (len(t) / 2)
    phase = np.angle(imag_ampl_out[i])
    restored_signal_out += ampli*np.cos(2*np.pi*freqs_out[i]*t_extend/100 + phase)
    ampli_in = np.absolute(imag_ampl_in[i]) / (len(t) / 2)
    phase_in = np.angle(imag_ampl_in[i])
    restored_signal_in += ampli_in*np.cos(2*np.pi*freqs_in[i]*t_extend/100 + phase_in)

```

```python
plt.figure(5)
plt.plot(t_extend, restored_signal_out, "r",  label=' out of phase' )
plt.plot(t_extend, restored_signal_in, "g", label = " in phase " )
plt.show()
```
